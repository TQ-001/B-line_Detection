This file converts the format of the pre-trained weight file to the format that can be used in fine-tuning. 

**Convert**

```python
cd /content/drive/MyDrive/DetCo/annotations/  # weight file location
```

```python
import torch

detco_radon = torch.load('/content/drive/MyDrive/DetCo/annotations/model_best.pth.tar')

print(detco_radon.keys())
```

```python
weight = detco_radon['state_dict']
```

```python
weight.keys()
```

```python
# remove 'module.'

weight_dict = {}

for k,v in weight.items():
  new_k = k.replace('module.','') if 'module' in k else k
  weight_dict[new_k] = v
```

```python
weight_dict.keys()
```

```python
# remove 'encoder_k.'

weight_dict_q = {}

for k,v in weight_dict.items():
  if 'encoder_q' in k:
    new_k = k.replace('encoder_q.','')  
    weight_dict_q[new_k] = v
    # print(len(weight_dict_q.keys()))
  else:
    continue
  
```

```python
weight_dict_q.keys()
```

```python
# remove repeat keys and mlps

weight_dict_q_resnet50 = {}

for k,v in weight_dict_q.items():
  if 'mlp' in k or 'layers' in k:
    continue
  else:
    weight_dict_q_resnet50[k] = v

# print(len(weight_dict_q_resnet50.keys()))
  
```

```python
weight_dict_q_resnet50.keys()
```

```python
detco_resnet50 = {"state_dict":weight_dict_q_resnet50,"author":"OpenSelfSup"}
```
